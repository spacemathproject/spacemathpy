```python
def PlotRtauSubplots(file,ncol,name_col={0:'X',1:'Y',2:'Z',3:'W',4:'R'},name=None):
    '''
    file: str
        File name with Rtau allowed parameter space
    ncol: int
        file has n columns, then ncol represent the column 
        that will represent axis x
    name_col: dict
        name_col is a dictionary with the names of every 
        column by default:
            
    
    usage:
        >>> PlotRtau('Rtau_random2sigma.txt',ncol=0)
    '''
    datos = np.loadtxt(file,delimiter=',')
    columnas = list(range(5))
    columnas.pop(ncol)
    fig,axes = plt.subplots(2,2,sharey=True)
    fig.figsize=(15,12)
    #fig.suptitle(f'${name_col[columnas[0]]}, {name_col[columnas[1]]}$'+
    #    f', ${name_col[columnas[2]]}, {name_col[columnas[3]]} vs {name_col[ncol]}$',y=1.05)
    axes[0,0].plot(datos[:,columnas[0]],datos[:,ncol],'.',color='#137A7A',alpha=0.5)
    axes[0,0].set(ylabel=fr'${name_col[ncol]}$', xlabel=f'${name_col[columnas[0]]}$')
    axes[0,1].plot(datos[:,columnas[1]],datos[:,ncol],'.',color='#7A137A',alpha=0.5)
    axes[0,1].set(xlabel=f'${name_col[columnas[1]]}$')
    axes[1,0].plot(datos[:,columnas[2]],datos[:,ncol],'.',color='#137A13',alpha=0.5)
    axes[1,0].set(ylabel=fr'${name_col[ncol]}$', xlabel=f'${name_col[columnas[2]]}$')
    axes[1,1].plot(datos[:,columnas[3]],datos[:,ncol],'.',color='#7A4713',alpha=0.5)
    axes[1,1].set(xlabel=f'${name_col[columnas[3]]}$')
    fig.tight_layout()
    if name==None:
        pass
    else:
        fig.savefig(name)
    #for ax in axes.flat:
    #    ax.label_outer()
```

```python
nombres_col = {0:'cos( - )',1:'tan()',2:'{{A_{tt}}}',3:'{{A_{bb}}}',4:'{{R_{}}}'}
PlotRtauSubplots('Rtau_random2sigma.txt',4,name_col=nombres_col,name='Rtau.png')
```

```python
def PlotSingle(file,colx,coly,name_col={0:'X',1:'Y',2:'Z',3:'W',4:'R'},color='#137A7A'):
    datos = np.loadtxt(file,delimiter=',')
    rows,colms = datos.shape
    minx,maxx = datos[:,colx].min(),datos[:,colx].max()
    miny,maxy = datos[:,coly].min(),datos[:,coly].max()
    x,y = (minx+maxx)/2.0,(miny+maxy)/2.0
    if colx in range(colms) and coly in range(colms):
        plt.figure()
        plt.plot(datos[:,colx],datos[:,coly],'o',color=color,alpha=0.5)
        plt.xlabel(fr'${name_col[colx]}$')
        plt.ylabel(fr'${name_col[coly]}$')
        plt.plot(x,y,'ro')
        #plt.grid()
    else:
        print(f'La columna debe ser un entero entre 0 y {colms-1}')
    
```

```python
PlotSingle('Rtau_random2sigma.txt',0,1,name_col=nombres_col)
```
